```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Heart.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.replace({'AHD':{'No': 0, 'Yes': 1}}, inplace=True)

df.replace({'Thal':{'fixed': 0, 'normal': 1, 'reversable': 2}}, inplace=True)

df.replace({'ChestPain':{'typical': 0, 'asymptomatic': 1, 'nonanginal': 2, 'nontypical': 3}}, inplace=True)
```

```python
df.tail()
```

```python
df['AHD'].value_counts()
```

```python
X = df.drop(columns='AHD', axis=1)
Y = df['AHD']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score on training data: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score on test data: ", test_data_accuracy)
```

```python
input_data = (5, 41, 0, 3, 130, 204, 0, 2, 172, 0, 1.4, 1, 0.0, 1 )

imput_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = imput_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==0):
    print("The person does not have a heart disease")
else:
    print("The person has a heart disease")
```
